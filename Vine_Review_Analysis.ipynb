```python
# Importing dependencies
import pandas as pd 
```

```python
# Reading in the vine_table.csv file that we originally uploaded into pgAdmin
vine_table_df = pd.read_csv("vine_table.csv")

# Viewing the first 10 rows of the vine_table DataFrame
vine_table_df.head(10)
```

```python
# Filtering the vine_table_df to keep only the rows where total_votes is greater than or equal to 20
filtered_votes_df = vine_table_df.loc[vine_table_df['total_votes'] >= 20]

# Showing the first 10 rows of the new DataFrame
filtered_votes_df.head(10)
```

```python
# Filtering filtered_votes_df to keep only the rows in which the number of helpful_votes divded by total_votes
# is greater than or equal to 50% (0.5)
helpful_total_ratio = filtered_votes_df.loc[(filtered_votes_df['helpful_votes'] / filtered_votes_df['total_votes']) >= 0.5]

# Showing the first rows of the new DataFrame
helpful_total_ratio.head(50)
```

```python
# Filtering help_total_ratio to keep only the rows in which the vine column has a value of "Y"
paid_reviews_df = helpful_total_ratio.loc[helpful_total_ratio['vine'] == "Y"]

# Showing the first 10 rows of written_reviews_df
paid_reviews_df.head(10)
```

```python
# Filtering helpful_total_ratio to keep only the rows in which the vine column has a value of "N"
unpaid_reviews_df = helpful_total_ratio.loc[helpful_total_ratio['vine'] == "N"]
unpaid_reviews_df.head(10)
```

```python
# Determining the total number of paid/unpaid Amazon reviews 
total_paid_reviews = paid_reviews_df["review_id"].count()
total_unpaid_reviews = unpaid_reviews_df["review_id"].count()

print("Total number of paid Amazon reviews: " + str(total_paid_reviews))
print("Total number of unpaid Amazon reviews: " + str(total_unpaid_reviews))
```

```python
# Determining the number of paid/unpaid 5 star reviews
paid_5_star = paid_reviews_df.loc[paid_reviews_df['star_rating'] == 5]['star_rating'].count()
unpaid_5_star = unpaid_reviews_df.loc[unpaid_reviews_df['star_rating'] == 5]['star_rating'].count()

print("Number of paid five star reviews: " + str(paid_5_star))
print("Number of unpaid five star reviews : " + str(unpaid_5_star))
```

```python
# Determining the percentage of paid 5 star reviews to total paid reviews
prct_of_paid = (paid_5_star / total_paid_reviews) * 100
print("Percentage of paid five star reviews to total paid reviews: " + str(prct_of_paid) + "%")

# Determining the percentage of unpaid 5 star reviews to total unpaid reviews
prct_of_unpaid = (unpaid_5_star / total_unpaid_reviews) * 100
print("Percentage of unpaid five star reviews to total unpaid reviews: " + str(prct_of_unpaid) + "%")
```
